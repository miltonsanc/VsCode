##(1) PLANTEAMIENTO DEL PROBLEMA: Información Genética para los organismos vivos

El MinTic en su grupo de investigación de Genética requiere un software que le permita gestionar adecuadamente la información obtenida del ADN, como se describe a continuación:

El ADN es la molécula de doble cadena que codifica la información genética para los organismos vivos. El ADN se compone de cuatro tipos de nucleótidos, que son moléculas que se unen entre sí para formar secuencias de ADN.

Los cuatro nucleótidos son adenina (A), guanina (G), citosina (C), y timina (T). Cada hebra de ADN es una secuencia de nucleótidos, por ejemplo AGCTAC. En el software se utilizará una representación de cadena de éste, "AGCTAC".

El ADN tiene 2 vertientes en una doble hélice. Los nucleótidos en una hebra se unen a los nucleótidos de la otra cadena. A y T pueden ser unidos entre sí, y por lo tanto se complementan entre sí, de manera similar, C y G son complementarios el uno del otro.

Puede ver ésto en [la página de Wikipedia para el ADN](http://en.wikipedia.org/wiki/Dna). Las dos cadenas de ADN son complementarias ya que cada nucleótido en una cadena está unido con su complemento en la otra cadena. Por lo tanto, teniendo en cuenta la secuencia de ADN ACGTACG, su cadena complementaria es TGCATGC.

##(5) CODIFICACIÓN PYTHON

````python
import random
# random.randint(1,4) Genera un numero aleatorio entre 0 y 10
def generarNucleotido():
  # construir función que retorne un nucleotido generado aleatoriamente
  nucleotido = random.randint(1,4)
  if nucleotido == 1:
    nucleotido = 'A'
  elif nucleotido == 2:
    nucleotido = 'G'
  elif nucleotido == 3:
    nucleotido = 'T'
  else: 
    nucleotido = 'C'
  return nucleotido # A G T C
def crearSecuencia():
  # construir función que retorne una secuencia de ADN valida con longitud 
  # aleatoria entre 20 y 100
  longitud =  random.randint(20, 100) #20
  arn = ''
  for i in range(longitud):
    arn = arn + generarNucleotido()
  return arn
    
def mostrarCadenas(cadena1, cadena2):
  print("\nCadena ADN uno : " , cadena1)
  print("Cadena ADN dos : " , cadena2 , "\n")

def obtenerLongitud(cadena1, cadena2):
  print ("Longitud Cadena 1: ", len(cadena1))
  print ("Longitud Cadena 2: ", len(cadena2))

def main():
  opcion = 0
  cadena1 = ''
  cadena2 = ''
  while opcion != 4:
    print("\n:: ============ MANEJO DE ADN ============ ::")
    print('''
          1. Generar secuencias
          2. Imprimir Secuencias
          3. Longitud de Secuencias
          4. Salir ''')
    opcion = 0
    while opcion < 1 or opcion > 4:
      opcion = int(input('\nElija Opcion :'))
      print("\n")

    if opcion == 1:
      cadena1 = crearSecuencia()
      cadena2 = crearSecuencia()
      print("\n:: ==== Secuencias Generadas con éxito === :: \n")
      mostrarCadenas(cadena1, cadena2)
      print("\n:: ======================================= :: \n")
    
    elif opcion == 2:
      mostrarCadenas(cadena1, cadena2)
    
    elif opcion == 3:
      obtenerLongitud(cadena1, cadena2)
    else:
      pass
  print("\nGracias, vuelva pronto!")


main()
````
